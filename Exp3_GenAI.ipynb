# **Implementation of PDF Document Ingestion and Text Chunking for Retrieval-Augmented Generation (RAG)**

## Objectives:
1. To extract textual content from a PDF document.
2. To apply text chunking techniques to split large documents into smaller meaningful parts.

###Step 1: Install Libraries

**Concept: Environment Setup.**

###Step 2: Upload PDF

**Concept: Data Input.**
We'll upload a PDF document from your local machine, which serves as our raw data source. Once uploaded, the file content will be accessible in the Colab environment.

```python
from google.colab import files
uploaded = files.upload()
```

###Step 3: Extract Text from PDF

**Concept: PDF Ingestion.**
This code extracts all textual content from the uploaded PDF, converting it into a single string. After running, the first 500 characters of the extracted text will be displayed for verification.

```python
import fitz

pdf_file = list(uploaded.keys())[0]
doc = fitz.open(pdf_file)

text = ""
for page in doc:
    text += page.get_text()

print(text[:500])
```

###Step 4: Apply Text Chunking

**Concept: Text Chunking.**
We use `RecursiveCharacterTextSplitter` with a `chunk_size` of 500 characters and `chunk_overlap` of 50 to divide the text. This process breaks the large document into smaller, manageable `chunks`, and the total number of chunks will be printed.

```python
from langchain_text_splitters import RecursiveCharacterTextSplitter

splitter = RecursiveCharacterTextSplitter(
    chunk_size=500,
    chunk_overlap=50
)

chunks = splitter.split_text(text)

print("Total Chunks:", len(chunks))
```

###Step 5: View Sample Chunks

**Concept: Chunk Output.**
This step allows us to visually inspect the first two generated chunks. By examining these samples, we can verify the quality and structure of the text segmentation.

```python
for i, chunk in enumerate(chunks[:2]):
    print(f"Chunk {i+1}:\n", chunk)
    print("-" * 40)
```
